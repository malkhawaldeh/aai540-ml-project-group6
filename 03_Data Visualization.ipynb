```python
# Create a copy to work on, preserving the original DataFrame from Athena
df_cleaned = df_from_athena.copy()

print("--- Step 1: Dropping Columns with High Missing Value Percentage ---")
# Using the corrected, lowercase column names
cols_to_drop = ['weapon used cd', 'weapon desc', 'crm cd 4', 'crm cd 3', 'crm cd 2', 'cross street', 'mocodes']
df_cleaned.drop(columns=cols_to_drop, inplace=True, errors='ignore')
print(f"DataFrame shape after dropping columns: {df_cleaned.shape}")


print("\n--- Step 2: Imputing Missing Values ---")
# For each column that has missing values, we'll fill them.
# We are using the recommended syntax 'df[col] = df[col].fillna(value)' to avoid warnings.

# Impute categorical columns with the mode (most frequent value)
for col in ['premis cd', 'premis desc', 'vict sex', 'vict descent', 'status', 'status desc', 'crm cd 1', 'location']:
    if col in df_cleaned.columns and df_cleaned[col].isnull().any():
        mode_val = df_cleaned[col].mode()[0]
        df_cleaned[col] = df_cleaned[col].fillna(mode_val)
        print(f"Imputed '{col}' with mode: '{mode_val}'")

# Impute numerical columns with the median (robust to outliers)
for col in ['vict age', 'lat', 'lon']:
    if col in df_cleaned.columns and df_cleaned[col].isnull().any():
        median_val = df_cleaned[col].median()
        df_cleaned[col] = df_cleaned[col].fillna(median_val)
        print(f"Imputed '{col}' with median: {median_val}")


# --- Step 3: Final Verification ---
print("\n--- Verifying Missing Values After Final Imputation ---")
missing_values_final = df_cleaned.isnull().sum()
if missing_values_final.sum() == 0:
    print("\n No missing values remain in the cleaned DataFrame.")
else:
    print("\n Missing values still exist. Columns with missing values:")
    print(missing_values_final[missing_values_final > 0])

# Display info to confirm all columns are full
print("\nFinal DataFrame Info:")
df_cleaned.info()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
import plotly.express as px

# The 'df_cleaned' DataFrame should already be in memory from the previous cleaning step.
df = df_cleaned

# --- Create an interactive Folium map (using a sample for performance) ---
print("--- Generating Interactive Crime Map (Folium) ---")
# Use a smaller sample if the DataFrame is very large
sample_df = df.sample(n=min(1000, len(df)), random_state=42) 

# Note the use of lowercase 'lat' and 'lon'
la_map = folium.Map(location=[sample_df['lat'].mean(), sample_df['lon'].mean()], zoom_start=11)

for _, row in sample_df.iterrows():
    if pd.notnull(row['lat']) and pd.notnull(row['lon']):
        folium.CircleMarker(
            # Using corrected lowercase column names
            location=[row['lat'], row['lon']],
            radius=2,
            color='red',
            fill=True,
            fill_color='#ff3333',
            fill_opacity=0.6,
            # Using .get() with corrected names for safety
            tooltip=f"Crime: {row.get('crm cd desc', 'N/A')}<br>Area: {row.get('area name', 'N/A')}"
        ).add_to(la_map)

display(la_map)


# --- Crime frequency by Time of Day ---
if 'time occ' in df.columns:
    print("\n--- Generating Crime Frequency by Hour ---")
    # Coerce errors will turn non-numeric time into 'NaT' (Not a Time) which we can drop
    df['time_hour'] = pd.to_numeric(df['time occ'], errors='coerce') // 100
    df.dropna(subset=['time_hour'], inplace=True)
    df['time_hour'] = df['time_hour'].astype(int)

    plt.figure(figsize=(12, 6))
    sns.countplot(x='time_hour', data=df, palette='magma')
    plt.title('Crime Frequency by Time of Occurrence')
    plt.xlabel('Hour of the Day (24-hour format)')
    plt.ylabel('Number of Incidents')
    plt.show()

# --- Crime frequency by Day of the Week ---
if 'date occ' in df.columns:
    print("\n--- Generating Crime Frequency by Day of Week ---")
    try:
        df['date_occ_dt'] = pd.to_datetime(df['date occ'], errors='coerce')
        df['day_of_week'] = df['date_occ_dt'].dt.day_name()

        plt.figure(figsize=(10, 6))
        day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
        sns.countplot(x='day_of_week', data=df, order=day_order, palette='viridis')
        plt.title('Crime Frequency by Day of the Week')
        plt.xlabel('Day of the Week')
        plt.ylabel('Number of Incidents')
        plt.show()
    except Exception as e:
        print(f"An error occurred during Day of Week plot: {e}")

# --- Crime frequency by Month ---
if 'date occ' in df.columns:
    print("\n--- Generating Crime Frequency by Month ---")
    try:
        # We can reuse 'date_occ_dt' from the previous step if it exists
        if 'date_occ_dt' in df.columns:
            df['month'] = df['date_occ_dt'].dt.month

            plt.figure(figsize=(12, 6))
            sns.countplot(x='month', data=df, palette='viridis')
            plt.title('Crime Frequency by Month')
            plt.xlabel('Month')
            plt.ylabel('Number of Incidents')
            plt.show()
    except Exception as e:
        print(f"An error occurred during Month plot: {e}")
```
